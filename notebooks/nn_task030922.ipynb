```python
def sin_fun(x):
  return np.sin(2*x)
```

```python
# формируем датасеты
def make_datasets():
  x_list = list(np.arange(0, np.pi, 0.01))
  t_list = [sin_fun(x) for x in x_list]
  y_list = [(t + (0.02 * np.random.randn() + 0)) for t in t_list]


  # dataframe чтобы рисовать можно было легко
  real_df = pd.DataFrame()
  real_df['x'] = x_list
  real_df['y'] = y_list
  real_df['t'] = t_list
  return real_df, x_list, t_list, y_list 
```

```python
real_df, x_list, t_list, y_list = make_datasets()
```

```python
real_df.plot(x='x', y=['y', 't'])
```

```python
def prepare_datasets(X_train, y_train):
  X_train = np.asarray(X_train, dtype=np.float64)
  y_train = np.asarray(y_train, dtype=np.float64)

  x_scaler = preprocessing.MinMaxScaler().fit(X_train)
  X_scaled = x_scaler.transform(X_train)

  y_scaler = preprocessing.MinMaxScaler().fit(y_train)
  Y_scaled = y_scaler.transform(y_train)

  return X_scaled, Y_scaled
```

```python
X_train, y_train = prepare_datasets(x_list, y_list)
```

```python
def make_network():
  model = tf.keras.Sequential()
  model.add(tf.keras.layers.Dense(10, input_dim=1, activation='relu'))
  model.add(tf.keras.layers.Dense(20, activation='relu'))
  model.add(tf.keras.layers.Dense(1, activation='softmax'))
  return model
```

```python
model = make_network()
```

```python
model.compile(loss= tf.keras.losses.MeanSquaredError(),
              metrics=tf.keras.metrics.MeanSquaredError(name="mean_squared_error", dtype=None), 
              optimizer='adam')
```

```python
history = model.fit(X_train, y_train, epochs=100, batch_size=64)
```

```python
y_pred = model.predict(X_train)
```

```python
pred_df = pd.DataFrame()
pred_df['x'] = real_df['x']
pred_df['y_pred'] = y_pred
pred_df['t'] = real_df['t']
```
